```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assume df is your DataFrame containing the 'median' and 'major_category' columns
# For illustration, let's create a sample DataFrame
np.random.seed(0)
df = pd.DataFrame({
    'major_category': np.random.choice(['Engineering', 'Business', 'Arts'], size=300),
    'median': np.random.normal(5000, 2000, size=300)  # Random normals with mean 5000 and std dev 2000
})

# Visualize the distribution of all median incomes
plt.figure(figsize=(14, 6))

# Try binwidth of $1000
plt.subplot(1, 2, 1)
sns.histplot(df['median'], bins=np.arange(0, 12001, 1000), kde=False)
plt.title('Histogram of Median Incomes\nBinwidth: $1000')
plt.xlabel('Median Income ($)')
plt.ylabel('Count')

# Try binwidth of $5000
plt.subplot(1, 2, 2)
sns.histplot(df['median'], bins=np.arange(0, 12001, 5000), kde=False)
plt.title('Histogram of Median Incomes\nBinwidth: $5000')
plt.xlabel('Median Income ($)')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Based on the histograms, choose a binwidth (let's say we choose $1000 for this example)
chosen_binwidth = 1000

# Plot the distribution of median income faceted by major_category
plt.figure(figsize=(14, 6))
sns.histplot(data=df, x='median', hue='major_category', multiple='stack', bins=np.arange(0, 12001, chosen_binwidth), kde=False)
plt.title(f'Faceted Histogram of Median Incomes by Major Category\nBinwidth: ${chosen_binwidth}')
plt.xlabel('Median Income ($)')
plt.ylabel('Count')
plt.legend(title='Major Category')
plt.show()
```
